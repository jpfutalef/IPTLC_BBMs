```python
import tqdm
import time

# Simulate the input by running the opf
simulated_angles_warm_restart = []
print(f" ---------- Simulating with warm start every 10 steps ---------- ")
# Setup a timer to measure the time it takes to simulate in seconds
start = time.time()

bar = tqdm.tqdm(range(num_of_days))

for i in bar:
    # Set the loads
    load = input_portion[i, :28]
    pg.set_external_stimuli(load)

    # set the condition
    piGen = input_portion[i, 28:33]
    piLine = input_portion[i, 33:48]
    piTrafo = input_portion[i, 48:]

    pg.set_gen_status(piGen)
    pg.set_transmission_line_status(piLine)
    pg.set_trafo_status(piTrafo)

    # run the opf
    init = "flat" if i % 10 == 0 else "results"
    pp.runopp(pg, verbose=False, init=init)
    simulated_angles_warm_restart.append(pg.res_gen.va_degree.values[:2])

# Measure the time it took to simulate
end = time.time()
print(f" ---------- Simulated with warm start every 10 steps in {end-start:.2f} seconds ---------- ")
```
